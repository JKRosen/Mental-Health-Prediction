```python
#this function takes the name of an illness
#returns a dictionary
#purpose- load all files
import glob
import re
from collections import defaultdict
def load(ill):
    path = str(ill) + '_filtered/*.txt'
    d = defaultdict()
    for filepath in glob.iglob(path):
        filename = filepath
        title = re.search(r"([0-9]\-).*(\.txt)", str(filename)).group(0)
        title = title[2:-4]
#         i += 1
        with open(filename, 'r',encoding="utf8") as file:
            post = file.read().replace('\n', '')
            d[title] = post
    return d
        
#this function takes a dictionary of posts
#returns a counter
#purpose- count pronouns
import spacy,en_core_web_sm
from collections import Counter
nlp = en_core_web_sm.load()
def pronouncount(illnessdict):
    clist = []
    for p in illnessdict:
        clist.append(Counter(([token.pos_ for token in nlp(illnessdict[p])])))
    return clist
      
#this function takes a dictionary of posts
#returns a dictionary
#purpose- removes \xa0 ; splits *punctuationI* errors
def clean(d):
    for post in d:
        d[post] = re.sub('\xa0',' ',d[post])
        d[post] = re.sub('[\.!\?]I', ". I", d[post])
        d[post] = re.sub('Â',' ',d[post])
#         d[post] = re.sub(r"(\\)",' ',d[post])
        d[post] = d[post].replace("I(\\)\'m","")
    d2 = d
    return d2

#this function takes a dictionary of posts
#returns a list of lists
#purpose- tokenize posts
def tokenize(d):
#     nlp = spacy.load("en_core_web_sm")
    PTSDlist = []
    for post in d:
#         tokenized = []
        doc = nlp(d[post])
        tokenized = [token.orth_ for token in doc if not token.is_punct | token.is_space]
#         for token in doc:
#             tokenized.append(token)
        PTSDlist.append(tokenized)   
    return PTSDlist

#this function takes a list of lists
#returns an integer
#purpose- count instances of first-person pronoun "I"
def countI(l):
    Icount = 0
    for i in l:
        for j in i:
            if "I" in j or j == "i":
                Icount += 1
    return Icount

#this function takes a list of lists
#returns an integer
#purpose- calculate word count
def wordcounter(l):
    wc = 0
    for i in l:
        for j in i:
            wc+=1
    return wc
```

## PTSD

```python
PTSDd = load('PTSD')
PTSDd = clean(PTSDd)
PTSDtoken = tokenize(PTSDd)
posc = pronouncount(PTSDd) #get counts for all parts of speech
ic = countI(PTSDd) #get I/i count

prontotal = 0
for c in posc: #get total pronoun count
    prontotal += c['PRON']

Iperpost = ic/prontotal
Iperpost
```
